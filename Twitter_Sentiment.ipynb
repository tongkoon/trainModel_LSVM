**Load Data**

```python
import pandas as pd
path='/content/drive/MyDrive/CS 3 1/AI/Lab LSTM/dataset/training-twitter.csv'
df = pd.read_csv(path, encoding='latin-1',header=None)
df.head()
```

```python
df.drop(df.columns[1:5], axis=1, inplace=True)
df = df.rename(columns={0: 'sentiment',5:'text'})
df.head()
```

```python
df.info()
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment'].replace({0:'Negative',4:'Positive'},inplace=True)
df
```

**Clean Data**

```python
import re

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')


def clean_data(text):
  pattern = re.compile(r'@\S+|https?://\S+|www\.\S+|http?:\S+|[^A-Za-z\s]+')  #ลบตัวเลขออกด้วย
  pattern = pattern.sub(r' ', text)
  pattern = pattern.lower()
  return pattern

def remove_stopword(text):
  stop_words = set(stopwords.words('english'))
  pattern=text.split()
  pattern=[word for word in pattern if not word in stop_words]
  pattern=' '.join(pattern)
  return pattern
```

```python
df['clean_data'] = df['text'].apply(lambda x: clean_data(x))
df
```

```python
df['remove_stopword'] = df['clean_data'].apply(lambda x: remove_stopword(x))
df
```

**Split Data**

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df,test_size=0.2, random_state=123)

train_df.shape, test_df.shape
```

**สร้าง Vector ของคำ**

```python
import gensim
```

```python
documents = [text.split() for text in df.remove_stopword]
```

```python
w2v_model = gensim.models.word2vec.Word2Vec(vector_size=100, window=5, min_count=10, workers=8)
w2v_model.build_vocab(documents)
```

```python
words = w2v_model.wv.key_to_index
vocab_size = len(words)
print('Vocab size:',vocab_size)
```

```python
w2v_model.train(documents, total_examples=len(documents), epochs=10)
```

**Padding Sequence of input data train and test**

```python
from keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(tokenizer.texts_to_sequences(train_df.remove_stopword), maxlen=100)
X_train
```

```python
X_test = pad_sequences(tokenizer.texts_to_sequences(test_df.remove_stopword), maxlen=100)
X_test
```

**ดึง target ออกจาก train และ test**

```python
y_train=train_df.sentiment
y_train.head()
```

```python
y_test=test_df.sentiment
y_test.head()
```

**สร้าง one hot vector สำหรับ target**

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y_train=labelencoder.fit_transform(y_train)
y_test=labelencoder.fit_transform(y_test)
```

```python
y_train.shape,y_test.shape
```

```python
del test_df
del train_df
```

**จัดข้อมูล input เพื่อนำเข้าสู่ LSTM**

```python
import numpy as np
embedding_matrix = np.zeros((vocab_size,100))
for word, i in tokenizer.word_index.items():
  if word in w2v_model.wv:
    embedding_matrix[i] = w2v_model.wv[word]
print(embedding_matrix.shape)
```

**สร้าง input layer หรือ embedding layer**

```python
from keras.layers import Embedding
embedding_layer = Embedding(vocab_size,100,weights=[embedding_matrix], input_length=100,trainable=False)
```

**สร้าง LSTM**

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout

model = Sequential()
model.add(embedding_layer)
model.add(Dropout(0.5)) #สุ่มลบค่า vector บางส่วน 50%
model.add(LSTM(100,dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1,activation='sigmoid')) #fn กระตุ้นเหมาะกับการ classification
model.summary()
```
